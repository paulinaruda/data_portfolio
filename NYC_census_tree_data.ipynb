### Data Preparation

#### Data Cleaning

```python
# to the values that already exist I will add value of Not applicable (NaN)
tcs.health.value_counts()
```

```python
# I create a mask that will give me the rows where the status of the tree is stump or dead
mask = ((tcs.status == 'Stump') | (tcs.status == 'Dead'))
```

```python
# and I will be specfying the rows and columns that I want to mask
# the first one specifies the row and the second one the column (column may be filled, but we are already working on a dataset 
# that has only nan values, so we want to fill all these where there are null values)
# if it was just one column it would be: tcs.loc[mask, 'health'] = tcs.loc[mask, 'health'].fillna('Not Applicable')
# fill those with nan values

tcs.loc[mask ] = tcs.loc[mask].fillna('Not Applicable')
```

```python
# now I need to decife if we keep those values or if we delete those rows or inter/extra polate
```

```python
# I will fill those - interpolation
tcs.health.fillna('Good', inplace = True)
tcs.sidewalk.fillna('NoDamage', inplace = True)
tcs.problems.fillna('None', inplace = True)
tcs.spc_latin.fillna('Not Observed', inplace = True)
```

```python
# checking if all worked
tcs.isna().sum()
```

#### Dealing with outliers

```python
big_trees
```

```python
# I could cut those as this is not a big subset of trees, but I decide to keep them and see how 
tcs_no_outliers = tcs[(tcs.tree_dbh <= 60) & (tcs.stump_diam <= 60)]
tcs_no_outliers
```

```python
# creating subsets for dead or alive trees as they have different kinds of data
tcs_alive = tcs[tcs.status == 'Alive']
tcs_dead_stump = tcs[(tcs.status == 'Stump') | (tcs.status == 'Dead')]
```

```python
# grouping the alive trees by their specie 
stats_alive = tcs_alive.groupby('spc_latin').mean()
stats_alive
```

```python
# providing some stats about the min and max diameter and the values of the 25% and 75% in deviation
stats_alive2 = tcs_alive.groupby('spc_latin')['tree_dbh'].describe()
stats_alive2
```

```python
# leaving only 25 and 75%
stats_alive3 = tcs_alive.groupby('spc_latin')['tree_dbh'].describe().reset_index()[['spc_latin', '25%', '75%']]
stats_alive3
```

```python
# merging both dataframes - to be able to compare to the lower or higher values (cutoff)
tcs_alive = tcs_alive.merge(stats_alive3, on = 'spc_latin', how = 'left')
tcs_alive
```

```python
tcs_alive.columns
```

```python
# if he diameter is lower than the lower limit (25%), then make the tree diameter(tree_dbh) equal to the lower limit
# if the condition evaluates to true then change it to the lower limit value

mask = tcs_alive['tree_dbh'] < tcs_alive['25%']
tcs_alive.loc[mask, 'tree_dbh'] = tcs_alive['25%']
```

```python
# if he diameter is higher than the lower limit (75%), then make the tree diameter (tree_dbh) equal to the higher limit
# if the condition evaluates to true then change it to the higher limit value

mask = tcs_alive['tree_dbh'] > tcs_alive['75%']
tcs_alive.loc[mask, 'tree_dbh'] = tcs_alive['75%']
```
