### Description 
Ash Ketchum caught some Pokemon and wants to find the evolution chain, but he can't find it in his Pokedex. Ash wants to start the training with those Pokemon and he is asking for some help to find the evolution chain and the variations on which they evolve using the Poke API, whose documentation can be found here<br>https://challenges.hackajob.co/pokeapi/docs/v2/ (https://challenges.hackajob.co/pokeapi/docs/v2/).
You need to return a string containing an object json encoded {"name": "butterfree", "variations": []}, variations contains a list of similar objects. A pokemon can have multiple evolution variations and we need to find them all.
<br><br>
Note:<br>
	•	A possible API you can use is https://challenges.hackajob.co/pokeapi/api/v2/pokemon-species/{id or name}/, using the evolutionchain_ field where you can find the entire chain. Variations contains all possible evolutions, not only consecutive evolutions. For example, metapod is not the only variation for caterpie, butterfree is also a variation because it's the evolution from metapod.<br>
	•	Variations are in order, for example if the input provided is butterfree or caterpie then the output is exactly the same because they are part of the same evolution chain.<br>
	•	For simplicity we will consider that a pokemon can evolve in only one other pokemon. For pokemons where there are multiple evolutions in the evolutuon chain we will consider the first one only.<br>
	•	You can print the response in your code and then see it in the Debug section inside of each test case. You can use https://jsonformatter.curiousconcept.com/ (https://jsonformatter.curiousconcept.com/) in order to beautify the JSON and understand it better.<br><br>
INPUT<br>
string   pokemonName<br><br>
OUTPUT<br>
json string    evolutionChain<br><br>
EXAMPLE<br>
Input<br>
"butterfree"<br><br>
Output<br>
{"name":"caterpie","variations":[{"name":"metapod","variations":[{"name":"butterfree","variations":[]}]}]} <br>\
